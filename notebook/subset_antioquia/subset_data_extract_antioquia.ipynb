```python
# Cargar y filtrar datos de Consultas Interpartidistas para Antioquia
import pandas as pd
consultas_path = '../../data/raw_data_extract/consultas_interpartidistas/cn_candidatos_municipio.parquet'
consultas_df = pd.read_parquet(consultas_path)
consultas_antioquia_df = consultas_df[consultas_df['depto_nombre'].str.upper() == 'ANTIOQUIA']
consultas_antioquia_df.reset_index(drop=True, inplace=True)
consultas_antioquia_df.head()
```

```python
consultas_antioquia_df.info()
```

```python
consultas_antioquia_df['candidato_votos'].sum()
```

```python
resultado = consultas_antioquia_df.groupby('partido_nombre')['candidato_votos'].sum()
resultado
```

```python
base = "/Users/ahenaor/Documents/projects/scraping-elecciones-colombia-2026/data/raw_data_extract_subset_antioquia/"
consultas_antioquia_df.to_excel(f'{base}consultas_antioquia.xlsx', index=False, engine='openpyxl')
```

```python
# Cargar y filtrar datos de Senado de la República para Antioquia
senado_path = '../../data/raw_data_extract/senado/se_candidatos_municipio.parquet'
senado_df = pd.read_parquet(senado_path)
senado_antioquia_df = senado_df[senado_df['depto_nombre'].str.upper() == 'ANTIOQUIA']
senado_antioquia_df.reset_index(drop=True, inplace=True)
senado_antioquia_df.head()
```
